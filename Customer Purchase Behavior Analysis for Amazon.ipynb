```python
# 1. Customer_Name: Impute missing names with "Unknown"
df_customer['Customer_Name'].fillna('Unknown', inplace=True)

# 2. Age: Impute missing values with the median age
median_age = df_customer['Age'].median()
df_customer['Age'].fillna(median_age, inplace=True)

# 3. Purchase_Date: Ensure the data type is datetime
df_customer['Purchase_Date'] = pd.to_datetime(df_customer['Purchase_Date'])

# 4. Purchase_Amount: Impute missing values with the median purchase amount
median_purchase = df_customer['Purchase_Amount'].median()
df_customer['Purchase_Amount'].fillna(median_purchase, inplace=True)

# 5. Rating: Impute missing values with the mode rating
mode_rating = df_customer['Rating'].mode()[0]
df_customer['Rating'].fillna(mode_rating, inplace=True)

# 6. Drop Feedback_Comments: Drop this column due to too many missing values
#df_customer.drop(columns=['Feedback'], inplace=True)

# Final check for missing values and data types
print(df_customer.isnull().sum())
print(df_customer.info())
print(df_customer.duplicated().sum())
# Step 2: Ensure correct data types
df_customer['Purchase_Date'] = pd.to_datetime(df_customer['Purchase_Date'])

# Step 3: Remove duplicates
#df_customer.drop_duplicates(inplace=True)
```

```python

# Cleaning Product Information DataFrame (product_info_df)
# Step 1: Handle missing values for Stock_Level, Discount, and Rating
df_product['Stock_Level'].fillna(df_product['Stock_Level'].median(), inplace=True)
df_product['Discount'].fillna(df_product['Discount'].mean(), inplace=True)
df_product['Rating'].fillna(df_product['Rating'].mean(), inplace=True)

# Step 2: Ensure correct data types for Price and other numeric fields
df_product['Price'] = pd.to_numeric(df_product['Price'], errors='coerce')
df_product['Stock_Level'] = pd.to_numeric(df_product['Stock_Level'], errors='coerce')

#If product categories exist, fill missing prices based on the median price within the same category
# Assuming a 'category' column exists
#df['price'] = df.groupby('category')['price'].transform(lambda x: x.fillna(x.median()))
# Step 3: Remove duplicates
df_product.drop_duplicates(inplace=True)

# Cleaning Transaction Details DataFrame (transaction_details_df)
# Step 1: Handle missing values for Shipping_Date and Feedback_Rating
df_transaction['Shipping_Date'] = pd.to_datetime(df_transaction['Shipping_Date'], unit='ms', errors='coerce')
df_transaction['Return_Reason'].fillna('No Reason', inplace=True)
df_transaction['Feedback_Rating'].fillna(df_transaction['Feedback_Rating'].mean(), inplace=True)
#Fill with the median shipping date (for example)
median_shipping_date = df_transaction['Shipping_Date'].median()
df_transaction['Shipping_Date'] = df_transaction['Shipping_Date'].fillna(median_shipping_date)
mode_delivery_status = df_transaction['Delivery_Status'].mode()[0]
df_transaction['Delivery_Status'] = df_transaction['Delivery_Status'].fillna(mode_delivery_status)
# Step 2: Remove duplicates
df_transaction.drop_duplicates(inplace=True)

# Check for duplicate rows
print("Duplicate rows in Customer Demographics dataset:", df_customer.duplicated().sum())

print("Duplicate rows in Product Information dataset:", df_product.duplicated().sum())

print("Duplicate rows in Transaction Details dataset:", df_transaction.duplicated().sum())
print(df_transaction.info())
print(df_transaction.isnull().sum())
print(df_product.isnull().sum())
```

```python
# Perform a left join with both transaction and customer data to keep all products
merged_df = pd.merge(df_product, df_transaction, on='Product_ID', how='left')
merged_df = pd.merge(merged_df, df_customer, on='Product_ID', how='left', suffixes=('_transaction', '_customer'))

# Fill missing customer or transaction data with placeholders
merged_df['Customer_Name'].fillna('Unknown Customer', inplace=True)
merged_df['Purchase_Amount'].fillna(0, inplace=True)

# Remove duplicated Customer_ID columns by keeping the one from customer data
merged_df['Customer_ID'] = merged_df['Customer_ID_customer'].fillna(merged_df['Customer_ID_transaction'])

# Drop the extra Customer_ID columns and any other redundant columns
merged_df.drop(['Customer_ID_transaction', 'Customer_ID_customer'], axis=1, inplace=True)

# Rename columns to be more descriptive and consistent
merged_df.rename(columns={
    'Product_Name': 'Product Name',
    'Purchase_Amount': 'Purchase Amount',
    'Purchase_Date': 'Purchase Date',
    'Rating_x': 'Product Rating',
    'Rating_y': 'Customer Rating'
}, inplace=True)

merged_df.drop_duplicates(subset=['Product_ID', 'Customer_ID','Transaction_ID'], inplace=True)

# Display the cleaned DataFrame
merged_df.head(20)
```

```python
# Aggregate data to combine rows with the same Product_ID and Transaction_ID
aggregated_df = merged_df.groupby(['Product_ID', 'Transaction_ID']).agg({
    'Purchase Amount': 'sum',  # Summing up the purchase amount for the same Product_ID and Transaction_ID
    'Rating_customer': 'mean',          # Averaging ratings if applicable
    'Customer_ID': 'first',# Keeping the first customer ID or using another logic
    'Customer_Name': 'first'
    
}).reset_index()

aggregated_df
```

```python


# 1. Feature: Total Purchase Amount per Customer
# Calculate the total purchase amount for each customer
merged_df['Total_Purchase_Amount'] = merged_df.groupby('Customer_ID')['Purchase Amount'].transform(sum)
merged_df
```

```python
merged_df.columns
```

```python
# 2. Feature: Frequency of Returns
# Calculate the number of times a customer has returned products
merged_df['Return_Count'] = merged_df.groupby('Customer_ID')['Return_Status'].transform(lambda x: (x == 'Returned').sum())

# 3. Feature: Average Feedback Rating
# Calculate the average feedback rating for each customer
merged_df['Average_Feedback_Rating'] = merged_df.groupby('Customer_ID')['Feedback_Rating'].transform('mean')

# 4. Feature: Time Since Last Purchase
# First, ensure that Purchase_Date is in datetime format
merged_df['Purchase Date'] = pd.to_datetime(merged_df['Purchase Date'], errors='coerce')

# Calculate the difference between the current date and the most recent purchase date
merged_df['Days_Since_Last_Purchase'] = (pd.Timestamp.now() - merged_df.groupby('Customer_ID')['Purchase Date'].transform('max')).dt.days


```

```python
merged_df['Customer_Lifetime_Value'] = merged_df.groupby('Customer_ID')['Purchase Amount'].transform('sum')
merged_df
```

```python
#merged_df['Days_Since_Last_Purchase'] = (pd.Timestamp.now() - merged_df.groupby('Customer_ID')['Purchase Date'].transform('max')).dt.days
merged_df['Churn_Risk'] = merged_df['Days_Since_Last_Purchase'].apply(lambda x: 1 if x >merged_df['Days_Since_Last_Purchase'].mean()  else 0)
merged_df
```

```python
merged_df['Loyalty_Status'] = (merged_df.groupby('Customer_ID')['Transaction_ID'].transform('count') > 5) & (merged_df['Days_Since_Last_Purchase'] < merged_df['Days_Since_Last_Purchase'].mean() )
merged_df
```

```python
merged_df['Purchase_Frequency'] = merged_df.groupby('Customer_ID')['Transaction_ID'].transform('count')
merged_df
```

```python
merged_df['Return_Count'] = merged_df.groupby('Customer_ID')['Return_Status'].transform(lambda x: (x == 'Returned').sum())
merged_df['Return_Rate_percentage'] = merged_df['Return_Count'] / merged_df['Purchase_Frequency']*100
merged_df
```

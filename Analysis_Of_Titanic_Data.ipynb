**Data Cleaning**

```python

#impute the missing values in average columns with the average age
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Age'].isna().sum()
```

```python
#the mode gives the o/p in series but we need to pick the  mode value
mode_emb = df['Embarked'].mode()[0]
mode_emb
```

```python
#we need to impute the embarked column with the mode value .Alse using inplace parameter rather than using the column value itself
df['Embarked'].fillna(mode_emb,inplace=True)
```

```python
#check for the missing value
df['Embarked'].isna().sum()
```

```python
df.isna().sum()
```

```python
#there is two possible way either we can add unknown or we can drop .since the no. of missing values are more so its best we drop it
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.columns
```

```python
#check duplicate values
df.duplicated().sum()
```

```python
df.head()
```

```python
#replace 'S' & 'C' in embarked
df['Embarked'].replace({'S':'SouthHampton','C':'California'},inplace=True)
```

```python
df.head()
```

```python
df['Sex'].replace({'male':'M','female':'F'},inplace=True)
```

```python
df.head()
```

Now that we have completed the cleaning ..Its important that the Well prepared data is key to accurate & meaningful insights.

**Data Manipulation**

```python
#What is the avergae age of passengers on the titanic?
print("Avergae age of passengers on the titanic:",df['Age'].mean())
```

```python
#How much total Fare was paid by passengers?
print("Total Fare paid by passengers:",df['Fare'].sum())
```

```python
#Determine the family size of each passengers by adding the 'SibSp'(Siblings/Spouses) and 'Parch'(parents/Childrens)columns.
df['FamilySize']=df['SibSp']+df['Parch']
df.head()
```

```python
#Calculate the survival rate (percentage of passengers who survived).
survived_df=df[df['Survived']==1] #conditional based filtering
survived_df
```

```python
#also i can use it through loc method (survived_df.loc[:,'Survived'].sum()*100/df.shape[0])
survival_rate=survived_df.shape[0]/df.shape[0]*100
print("Survival Rate:",survival_rate)
```

```python
#who are the passengers under the age of 18?
df[df['Age']<18]
```

```python
#create a subset of data containing only the following columns:'Name','Age','Sex','Survived','FamilySize'.
df.loc[:,['Name','Age','Sex','Survived','FamilySize']] #label based slicing
```

```python
#Gender ratio
No_of_male=df[df['Sex']=='M']
No_of_female=df[df['Sex']=='F']
print("No. of male passengers:",No_of_male.shape[0])
print("No. of female passengers:",No_of_female.shape[0])
```

```python
#male & female percentage
male_perc=No_of_male.shape[0]/df.shape[0]*100
female_perc=No_of_female.shape[0]/df.shape[0]*100
print("Percentage of male passengers:",male_perc)
print("Percentage of female passengers:",female_perc)
```
